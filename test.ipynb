```python
import numpy as np
def cosine_similarity(x,y):
     cos_sim = np.dot(x,y)/(np.linalg.norm(x)*np.linalg.norm(y))
     return cos_sim
```

```python
model = Word2Vec.load("wordvectors\skill2skill2.model")
```

```python
## Function to process given array of skills to match those in model vocabulary
def process_input_skills(array, unique_words=model.wv.key_to_index):
    return [word for word in array if word in unique_words]
```

```python
from gensim.models import Word2Vec
```

```python
p_skills = clean_skills(proj_skills[0])
f_skills = []
for f in range(50):
    f_skills.append(skills[np.random.choice(len(skills),1)].values[0])
```

```python
p_skills_vec = model.wv[process_input_skills(p_skills)][:20]
f_skills_vec = model.wv[process_input_skills(f_skills)][:20]

sim_vec = cosine_similarity(f_skills_vec,p_skills_vec.T)
```

```python
## TRI SUM
tri_sum_sim_vec = (sim_vec.sum() - np.diag(sim_vec).sum())/2
tri_sum_sim_vec
```

```python
## AVERAGE
ave_sim_vec = np.average(sim_vec)
ave_sim_vec
```
